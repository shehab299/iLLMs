```python
import os
from groq import Groq
import json
from Tool import Tool
from Agent import Agent
```

```python
os.environ['GROK_API_KEY'] = "<YOUR_API_KEY>"
client = Groq(api_key=os.environ.get("GROK_API_KEY"))
```

#### Tools

```python
@Tool
def add(x: int, y: int) -> int:
    """Adds two numbers."""
    return x + y

@Tool
def multiply(x: int, y: int) -> int:
    """Multiplies two numbers."""
    return x * y

@Tool
def subtract(x: int, y: int) -> int:
    """Subtracts two numbers."""
    return x - y
```

```python
shosho = Agent(
    name="shosho",
    role="A helpful assistant that can perform basic arithmetic operations.",
    tools=[add, multiply, subtract],
    client=client,
    system_message_file="SYSTEM_PROMPT.txt",
)
```

```python
shosho("What is the sum of 5 and 3? Can you also multiply 4 and 2, and then subtract 1 from the result?")
```

```python
@Tool
def getTodayDate() -> str:
    """Returns today's date."""
    from datetime import datetime
    return datetime.now().strftime("%Y-%m-%d")
```

```python
soso = Agent(
    name="soso",
    role="A helpful assistant.",
    tools=[getTodayDate],
    client=client,
    system_message_file="SYSTEM_PROMPT.txt",
)
```

```python
soso("Can you tell me today's date?")
```
